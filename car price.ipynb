```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Importing the dataset
dataset = pd.read_csv('D:\ML\A3 Data\CarPrice_Assignment.csv')
# Cheking the null value 
print(dataset.isnull().sum())

# Show a head of data 
dataset.head() 
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset.iloc[:,2] = le.fit_transform(dataset.iloc[:,2])
dataset.iloc[:,3] = le.fit_transform(dataset.iloc[:,3])
dataset.iloc[:,4] = le.fit_transform(dataset.iloc[:,4])
dataset.iloc[:,5] = le.fit_transform(dataset.iloc[:,5])
dataset.iloc[:,6] = le.fit_transform(dataset.iloc[:,6])
dataset.iloc[:,7] = le.fit_transform(dataset.iloc[:,7])
dataset.iloc[:,8] = le.fit_transform(dataset.iloc[:,8])
dataset.iloc[:,14] = le.fit_transform(dataset.iloc[:,14])
dataset.iloc[:,15] = le.fit_transform(dataset.iloc[:,15])
dataset.iloc[:,17] = le.fit_transform(dataset.iloc[:,17])

dataset.head() 
```

```python
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values




  # Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler() # take an object from StandardScaler 
X_train = sc.fit_transform(X_train) # convert X_train 
X_test = sc.transform(X_test) # convert X_test
```

```python
from sklearn.linear_model import LinearRegression
regressor1_MR = LinearRegression() # take an object from class LinearRegression
regressor1_MR.fit(X_train, y_train) # fit the model in train data

# Predicting the Test and Train set results
y_pred_MR = regressor1_MR.predict(X_test)


y_pred_train_MR = regressor1_MR.predict(X_train)
```

```python
from sklearn import metrics 
# calculate and print the Accuracy in train 
acc_train1 = metrics.r2_score(y_train , y_pred_train_MR)
print("Accuracy In Train = ",acc_train1)



# calculate and print the Accuracy in test 
acc_test1 = metrics.r2_score(y_test , y_pred_MR)
print("Accuracy In Test = ",acc_test1)
```

```python
# Support Vector Machine
from sklearn.svm import SVR 
regressor2_SV = SVR(kernel = 'rbf') # we will use kernel in rbf tybe
regressor2_SV.fit(X_train,y_train) # fit the model in train data

# Predicting the Test and Train set results
y_train_pred_SV = regressor2_SV.predict(X_train) 
y_pred_SV = regressor2_SV.predict(X_test)

```

```python
# calculate and print the Accuracy in train

acc_train_SV = metrics.r2_score(y_train,y_train_pred_SV)
print("Accuracy In Train = ",acc_train_SV)

# calculate and print the Accuracy in test
acc_test_SV = metrics.r2_score(y_test,y_pred_SV)
print("Accuracy In Test = ",acc_test_SV)
```

```python

# Training the Multiple Linear Regression model on the Training set
from sklearn.ensemble import RandomForestRegressor
regressor_rf = RandomForestRegressor(n_estimators = 1500, random_state = 0)# use 1500  random forest regressor
regressor_rf.fit(X_train, y_train) # fit the model in train data

# Predicting the Test and Train set results
y_pred_RF  = regressor_rf.predict(X_test)
y_pred_train_RF = regressor_rf.predict(X_train)
```

```python
# calculate and print the Accuracy in train

acc_train_RF = metrics.r2_score(y_train , y_pred_train_RF)
print("Accuracy In Train = ",acc_train_RF)


# calculate and print the Accuracy in test

acc_test_RF = metrics.r2_score(y_test , y_pred_RF)
print("Accuracy In Test = ",acc_test_RF)
```

```python
# Training the Decision Tree Regression model on the whole dataset
from sklearn.tree import DecisionTreeRegressor
regressor2_DS = DecisionTreeRegressor(random_state = 0) # make an object feom class DecisionTreeRegressor
regressor2_DS.fit(X_train, y_train) # Fit the model in train data 

# Predicting the Test and Train set results
y_train_pred_DS = regressor2_DS.predict(X_train) 
y_pred_DS = regressor2_DS.predict(X_test) 
```

```python
# calculate and print the Accuracy in train

acc_train_DS = metrics.r2_score(y_train,y_train_pred_DS)
print("Accuracy In Train = ",acc_train_DS)

# calculate and print the Accuracy in test
acc_test_DS = metrics.r2_score(y_test,y_pred_DS)
print("Accuracy In Test = ",acc_test_DS)
```
